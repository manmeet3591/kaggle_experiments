```python
# Creating custom Dataset classes

class MyDataset(torch.utils.data.Dataset):
    def __init__(self, data, target, transform=None):
        # 1. Initialize file paths or a list of file names. 
        self.data = torch.from_numpy(data).float()
        self.target = torch.from_numpy(target).long()
        self.transform = transform
        
    def __getitem__(self, index):
        # 1. Read one data from file (e.g. using numpy.fromfile, PIL.Image.open).
        # 2. Preprocess the data (e.g. torchvision.Transform).
        # 3. Return a data pair (e.g. image and label).
        x = self.data[index]
        y = self.target[index]
        
        if self.transform:
            x = self.transform(x)
            
        return x, y
    
    def __len__(self):
        # Total size of dataset
        return len(self.data)
```

```python
numpy_data = np.random.randn(100,3,224,224) # 10 samples, image size = 224 x 224 x 3
numpy_target = np.random.randint(0,5,size=(100))
```

```python
print(numpy_data.shape, numpy_target.shape)
```

```python
dataset = MyDataset(numpy_data, numpy_target)
print(dataset)
```

```python
# Hyper parameters
num_epochs = 5
num_classes = 10
batch_size = 100
learning_rate = 0.001
```

```python
# Data loader
loader = torch.utils.data.DataLoader(dataset=dataset,
                                           batch_size=batch_size, 
                                           shuffle=True)
```

```python
# Device configuration
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```
